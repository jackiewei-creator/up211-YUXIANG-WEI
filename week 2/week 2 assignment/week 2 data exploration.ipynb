# Import the data

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import the data
data_path = '/mnt/data/stay point .csv'  # Adjust the file path if necessary
data = pd.read_csv(data_path)
```

# Data Exploration

```python
# Check the shape of the data (number of rows and columns)
data.shape
```

```python
# Display information about the dataset (columns, non-null counts, and data types)
data.info()
```

```python
# Display the first few rows of the dataset
data.head()
```

# Visualize Data Distribution

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['longitude'], data['latitude'], alpha=0.5)
plt.title('Scatter Plot of Latitude vs Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

# Count Unique Values

```python
activity_counts = data['activity_type'].value_counts()
print("Activity Type Counts:")
print(activity_counts)

# Create a bar chart for activity counts
plt.figure(figsize=(8, 6))
activity_counts.plot(kind='bar')
plt.title('Activity Type Counts')
plt.xlabel('Activity Type')
plt.ylabel('Frequency')
plt.show()
```
