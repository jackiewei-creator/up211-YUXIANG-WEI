# Step 1 - Data Loading and Initial Exploration
In this step, I will load the dataset and perform an initial inspection.
I will check the first five rows, dataset dimensions, column information, and missing values.

```python
import pandas as pd

# Load the dataset
file_path = "ACSST5Y2023.S1903-Data.csv"
data = pd.read_csv(file_path)

# Display the first five rows
print("First five rows of the dataset:")
data.head()
```

```python
# Display information about the dataset
print("\nDataset info:")
data.info()
```

```python
# Check for missing values
print("\nMissing values in each column:")
print(data.isnull().sum())
```

# Step 2 - Data Cleaning and Selection
In this step, we will extract three important columns:

```python
# Select relevant columns
data_cleaned = data[["GEO_ID", "NAME", "S1903_C03_015E"]].copy()

# Display the first five rows of the selected data
print("\nSelected relevant columns (first five rows):")
data_cleaned.head()
```

```python
# Skip rows with non-numeric values in Medium_Income_Households
data_cleaned = data[["GEO_ID", "NAME", "S1903_C03_015E"]].copy()

# Rename columns
data_cleaned.columns = ["Census_Tract_ID", "Census_Tract_Name", "Medium_Income_Households"]

# Remove rows where Medium_Income_Households is non-numeric
data_cleaned = data_cleaned[pd.to_numeric(data_cleaned["Medium_Income_Households"], errors="coerce").notnull()]

# Convert Medium_Income_Households to float
data_cleaned["Medium_Income_Households"] = data_cleaned["Medium_Income_Households"].astype(float)

# Check the cleaned data
print("\nCleaned dataset shape:", data_cleaned.shape)
```

```python
data_cleaned.head()
```

# Step 3 - Data Type Conversion and Cleaning
I will handle invalid values in the `Total_Households` column by replacing "**" with NaN

```python
# Replace invalid values and convert to float
data_cleaned["Medium_Income_Households"] = data_cleaned["Medium_Income_Households"].replace("**", None).astype(float)

# Display the data types and the first five rows after conversion
print("\nData types and first five rows after conversion:")
print(data_cleaned.dtypes)

# Drop rows with missing values in the Medium_Income_Households column
data_cleaned = data_cleaned.dropna(subset=["Medium_Income_Households"])

# Display the shape of the cleaned dataset
print("\nCleaned dataset shape (rows, columns):", data_cleaned.shape)
```

# Step 4 - Categorizing Data
Based on the `Medium_Income_Households` column, we will categorize Census Tracts into three groups:
Low Income: Less than $50,000
Middle Income: Between $50,000 and $100,000
High Income: More than $100,000

```python
# Define the categorization function
def categorize_income(income):
    if income < 50000:
        return "Low Income"
    elif income <= 100000:
        return "Middle Income"
    else:
        return "High Income"

# Apply the function to create a new column for income category
data_cleaned["Income_Category"] = data_cleaned["Medium_Income_Households"].apply(categorize_income)

# Display the first five rows after categorization
print("\nFirst five rows after categorization:")
data_cleaned.head()
```

# Step 5 - Statistical Analysis
I will calculate the number of Census Tracts in each income category and their percentage of the total.

```python
# Count the number of Census Tracts in each category
income_counts = data_cleaned["Income_Category"].value_counts()
income_percentage = income_counts / income_counts.sum() * 100

# Display the results
print("\nNumber of Census Tracts by income category:")
print(income_counts)
print("\nPercentage of Census Tracts by income category:")
print(income_percentage)
```

# Step 6 - Bar chart 
I will create a bar chart to display the distribution of Census Tracts by income category.

```python
import matplotlib.pyplot as plt

# Create a bar chart
fig, ax = plt.subplots(figsize=(8, 6))
income_counts.plot(kind="bar", color=["red", "orange", "green"], alpha=0.7, ax=ax)
ax.set_title("Number of Census Tracts by Income Level", fontsize=14)
ax.set_xlabel("Income Category", fontsize=12)
ax.set_ylabel("Number of Census Tracts", fontsize=12)

# Add percentage labels on top of each bar
for i, (count, pct) in enumerate(zip(income_counts, income_percentage)):
    ax.text(i, count + 2, f"{pct:.1f}%", ha="center", fontsize=10)

plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import re

# Step 1: Print the original data to inspect the format of the Census_Tract_Name column
print("Original Census_Tract_Name:")
print(data_cleaned["Census_Tract_Name"].head(10))

# Step 2: Define a function to extract numeric parts with decimals from Census_Tract_Name
def extract_numeric_with_decimal(name):
    """
    Extract numeric parts (including decimals) from the Census_Tract_Name field.
    Example: 'Census Tract 1011.10; Los Angeles County; California' -> '1011.10'
    """
    if pd.isnull(name):  # Return None if the value is null
        return None
    # Match the numeric part after "Census Tract"
    match = re.search(r"Census Tract (\d+\.\d+|\d+)", name)
    return match.group(1) if match else None

# Step 3: Apply the extraction function to the Census_Tract_Name column
data_cleaned["Census_Tract_Name"] = data_cleaned["Census_Tract_Name"].apply(extract_numeric_with_decimal)

# Step 4: Print the extracted values to confirm the results
print("Extracted Census_Tract_Name:")
print(data_cleaned["Census_Tract_Name"].head(10))
```
